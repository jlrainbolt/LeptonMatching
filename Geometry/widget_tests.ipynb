```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.autoscale(enable=False)

elec1 = rt.TLorentzVector(-1.1588054941, -4.5484023317, -4.4258494716, 6.4512740982)
elec2 = rt.TLorentzVector(-5.0590000905, -21.994286557, -8.2352653512, 24.024190277)
elec3 = rt.TLorentzVector(-8.0841867049, 36.466366777, -2.7641945719, 37.453848297)
elec4 = rt.TLorentzVector(14.301992290, -9.9236778883, 15.425309395, 23.258687328)
elecs = [elec1, elec2, elec3, elec4]
col = {elec1:'b', elec2:'r', elec3:'b', elec4:'r'}

for elec in elecs:
    ax.quiver(0, 0, 0, elec.Px(), elec.Py(), elec.Pz(), pivot='tail', length=elec.P(), color=col[elec])

lim=40
ax.set_xlim(left=-10, right=15)
ax.set_ylim(bottom=-23, top=37)
ax.set_zlim(bottom=-9, top=16)
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.autoscale(enable=False)

elec1 = rt.TLorentzVector(+3.9097633371e+00, +8.9282537853e+00, -2.2133446675e+01, 2.4184487316e+01)
elec2 = rt.TLorentzVector(+2.0955508463e+00, +6.7216496357e+00, +3.7627105108e+01, 3.8280163871e+01)
muon1 = rt.TLorentzVector(+5.8902973312e+00, -3.5136765516e+00, -2.3867035725e+00, 7.2620850657e+00)
muon2 = rt.TLorentzVector(-1.1895611515e+01, -1.2136226869e+01, -1.3106954861e+01, 2.1461263748e+01)
leps = [elec1, elec2, muon1, muon2]
col = {elec1:'r', elec2:'r', muon1:'b', muon2:'b'}
q = {elec1:'_', elec2:'+', muon1:'_', muon2:'+'}

for lep in leps:
    ax.quiver(0, 0, 0, lep.Px(), lep.Py(), lep.Pz(), pivot='tail', length=lep.P(), color=col[lep], arrow_length_ratio=2/lep.P())
    ax.scatter(lep.Px(), lep.Py(), lep.Pz(), marker=q[lep], s=200, c='k')

lim=20
ax.set_xlim(left=-lim, right=lim)
ax.set_ylim(bottom=-lim, top=lim)
ax.set_zlim(bottom=-lim, top=lim)
plt.show()
```
